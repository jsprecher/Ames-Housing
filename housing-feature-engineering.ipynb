# Feature Engineering

```python
import pandas as pd
import numpy as np
import sklearn as sk
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import *
import scipy.stats as stats
```

```python
train = pd.read_csv('train.csv')
train.columns
```

```python
train['total_SF'] = train[['TotalBsmtSF','1stFlrSF','2ndFlrSF']].sum(axis=1)
```

```python
train.loc[train['MoSold'].apply(lambda x: x in [12,1,2]),'Season'] = "winter"
train.loc[train['MoSold'].apply(lambda x: x in [3,4,5]),'Season'] = "spring" 
train.loc[train['MoSold'].apply(lambda x: x in [6,7,8]),'Season'] = "summer"
train.loc[train['MoSold'].apply(lambda x: x in [9,10,11]),'Season'] = "fall"

sns.boxplot(x=train['OverallQual'], y=train['SalePrice'])
plt.show()

sns.boxplot(x=train['Season'], y=train['SalePrice'])
```
